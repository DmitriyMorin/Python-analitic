## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix
```

```python
corr_matrix = np.round(corr_matrix, 1)
corr_matrix
```

```python
corr_matrix[np.abs(corr_matrix)<=0.5] = 0
corr_matrix
```

```python
upp_mat = np.triu(df.corr())
```

```python
plt.figure(figsize = (20, 5))
sns.heatmap(df.corr(), annot=True, cmap = 'viridis', mask = upp_mat);
```

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg')
plt.xlabel('Квадратура жилой площади')
plt.ylabel('Цена')
plt.grid();
```

Максимальная корреляция в 70% достигается в сзязке Цена/Квадратура жилой площади;
Большая часть домов имеет жилую площадь до 6 тысяч и стоимость 2 млн.

```python
sns.jointplot(x=df['grade'], y=df['price'], kind='kde')
plt.xlabel('Класс дома')
plt.ylabel('Цена')
plt.grid();
```

Корреляция 67% имеет связку Цена/Класс
Большая часть приходится с 6 по 8 Класс

```python
sns.jointplot(x=df['sqft_above'], y=df['price'], kind='reg')
plt.xlabel('Квадратура общей площади дома')
plt.ylabel('Цена')
plt.grid();
```

Корреляция в связке Цена/Квадратура общей площади составляет 61%
Данные схожи с Цена/Квадратура жилой площади. Аномальных откланений нет.

```python
sns.boxplot(x=df['view'], y=df['price'], whis=1.5)
plt.xlabel('Вид')
plt.ylabel('Цена');
```
